# TRATANDO DADOS AUSENTES

```python
#verificando dados nulos
#true = dados faltantes
#false = dados não faltantes nas colunas
df.isnull()
```

```python
#verificando dados nulos para cada coluna
df.isnull().sum()
```

```python
#excluindo os dados ausentes
dado = df.dropna()
dado.shape
```

```python
dado.isnull().sum()
```

```python
#dados ausentes percentualmente
faltantes_percentual = (dado.isnull().sum() / len(dado['_id']))*100
print(faltantes_percentual)
```

```python
# Excluir as colunas "Age Group" e "Reported Date"
dado = dado.drop(columns=['Age Group', 'Reported Date', '_id'])
print(dado.columns)
```

```python
# Transformando strings 'yes' e 'no' para numérico
cols_to_convert = [
    'Currently Hospitalized',
    'Currently in ICU',
    'Currently Intubated',
    'Ever Hospitalized',
    'Ever in ICU',
    'Ever Intubated'
]

# Mapeie 'yes' para 1 e 'no' para 0 em todas as colunas especificadas no DataFrame 'dado'
dado.loc[:, cols_to_convert] = dado.loc[:, cols_to_convert].replace({'Yes': 1, 'No': 0})
print(dado.head())
```

```python
# invertendo os valores das colunas strings
dado.loc[:, 'Classification'] = dado['Classification'].replace({'CONFIRMED': 1, 'PROBABLE': 0})
dado.loc[:, 'Client Gender'] = dado['Client Gender'].replace({'MALE': 1, 'FEMALE': 0})
dado.loc[:, 'Outcome'] = dado['Outcome'].replace({'RESOLVED': 1, 'FATAL': 0})
dado.head()
```

```python
dado.to_csv("pre-processed_data_COVID19.csv")
```
